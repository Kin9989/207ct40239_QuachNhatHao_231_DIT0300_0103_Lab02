```python
# lab 2 cu 1
import cv2
import matplotlib.pyplot as hao

# c nh
image_path = "/content/Screenshot 2023-11-05 193304.png"
original_image = cv2.imread(image_path)

# Lm m nh s dng cv2.blur()
blurred_image_cv2 = cv2.blur(original_image, (5, 5))  # Chn kch thc filter (5x5)

# Lm m nh s dng cv2.boxFilter()
box_filtered_image = cv2.boxFilter(original_image, -1, (5, 5))  # Chn kch thc filter (5x5)

# Biu din nh gc v nh lm m cng nhau
hao.subplot(1, 3, 1)
hao.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
hao.title('nh Gc')

hao.subplot(1, 3, 2)
hao.imshow(cv2.cvtColor(blurred_image_cv2, cv2.COLOR_BGR2RGB))
hao.title('nh Lm M (cv2.blur())')

hao.subplot(1, 3, 3)
hao.imshow(cv2.cvtColor(box_filtered_image, cv2.COLOR_BGR2RGB))
hao.title('nh Lm M (cv2.boxFilter())')

hao.show()
```

```python
# cu 4 lab 2
import cv2
import matplotlib.pyplot as hao

# c hai nh
image1_path = "/content/Screenshot 2023-11-05 194917.png"
image2_path = "/content/Screenshot 2023-11-05 194922.png"

img1 = cv2.imread(image1_path)
img2 = cv2.imread(image2_path)

# p dng cc b lc
kernel_size = 5

# Box Filter
box_filtered_img1 = cv2.boxFilter(img1, -1, (kernel_size, kernel_size))
box_filtered_img2 = cv2.boxFilter(img2, -1, (kernel_size, kernel_size))

# Gaussian Filter
gaussian_blur_img1 = cv2.GaussianBlur(img1, (kernel_size, kernel_size), 0)
gaussian_blur_img2 = cv2.GaussianBlur(img2, (kernel_size, kernel_size), 0)

# Median Filter
median_blur_img1 = cv2.medianBlur(img1, kernel_size)
median_blur_img2 = cv2.medianBlur(img2, kernel_size)

# Biu din nh gc v cc nh lm m
hao.figure(figsize=(12, 8))

hao.subplot(2, 3, 1)
hao.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
hao.title('nh Gc 1')

hao.subplot(2, 3, 2)
hao.imshow(cv2.cvtColor(box_filtered_img1, cv2.COLOR_BGR2RGB))
hao.title('Box Filter 1')

hao.subplot(2, 3, 3)
hao.imshow(cv2.cvtColor(gaussian_blur_img1, cv2.COLOR_BGR2RGB))
hao.title('Gaussian Blur 1')

hao.subplot(2, 3, 4)
hao.imshow(cv2.cvtColor(median_blur_img1, cv2.COLOR_BGR2RGB))
hao.title('Median Blur 1')

hao.subplot(2, 3, 5)
hao.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
hao.title('nh Gc 2')

hao.subplot(2, 3, 2)
hao.imshow(cv2.cvtColor(box_filtered_img2, cv2.COLOR_BGR2RGB))
hao.title('Box Filter 2')

hao.subplot(2, 3, 3)
hao.imshow(cv2.cvtColor(gaussian_blur_img2, cv2.COLOR_BGR2RGB))
hao.title('Gaussian Blur 2')

hao.subplot(2, 3, 6)
hao.imshow(cv2.cvtColor(median_blur_img2, cv2.COLOR_BGR2RGB))
hao.title('Median Blur 2')

hao.show()
```

```python
# cu 5 lab 2
# Cc kch thc b lc (kernel size) ng vai tr quan trng trong vic x l nh v lm m hnh nh. Di y l mt s nhn xt v cc kch thc filter:

#     1 Kch Thc B Lc Ln:
#         Khi s dng b lc ln (v d: kernel size ln), hiu ng lm m s mnh m hn v lm gim chi tit trong hnh nh.
#         Tuy nhin, iu ny c th lm mt thng tin quan trng nu kch thc qu ln.

#     2 Kch Thc B Lc Nh:
#         B lc nh (v d: kernel size nh) s gi li nhiu chi tit hn trong hnh nh, nhng cng lm gim  m.
#         C th to ra hiu ng lm m nh v gi li nhiu chi tit cn thit.

#     3 Kernel Size L S L:
#         Trong mt s trng hp, c bit l khi s dng median filter, vic chn kernel size l s l thng c khuyn khch. iu ny gip gi li gi tr trung bnh ca pixel mt cch hiu qu hn.

#     4 La Chn Kernel Size Ph Hp:
#         La chn kch thc b lc ph thuc vo c tnh ca hnh nh v mc tiu ca qu trnh x l.
#         Cn th nghim v iu chnh kch thc b lc  t c hiu qu tt nht cho tng tnh hung c th.

#     5 Ty Chn Ti u:
#         Trong mt s trng hp, s kt hp ca cc b lc khc nhau hoc s thay i kch thc b lc theo tng giai on c th dn n kt qu tt hn.
```

```python
# cu 6 lab 2
import cv2
import matplotlib.pyplot as hao

# c nh v lm m
img = cv2.imread('/content/pepper_noise.jpg')
blur = cv2.medianBlur(img, 5)

# Chuyn nh trng en sang nh mu
color_img = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
# Chuyn nh trng en sang nh mu xanh
enhanced_green_img = color_img.copy()
enhanced_green_img[:, :, 1] = cv2.equalizeHist(color_img[:, :, 1])

# Chuyn nh mu sang nh en trng
gray_img = cv2.cvtColor(color_img, cv2.COLOR_RGB2GRAY)

# Cn bng sng vi equalizeHist
equalized_img = cv2.equalizeHist(gray_img)

# Biu din cc nh
hao.figure(figsize=(10, 8))

hao.subplot(2, 2, 1)
hao.imshow(enhanced_green_img)
hao.title('nh Mu Xanh')

hao.subplot(2, 2, 2)
hao.imshow(gray_img, cmap='gray')
hao.title('nh en Trng')

hao.subplot(2, 2, 3)
hao.imshow(equalized_img, cmap='gray')
hao.title('nh Cn Bng Sng')

hao.show()
```

```python
import cv2
import matplotlib.pyplot as hao

# Read the two images
img1 = cv2.imread('/content/exT3_01.jpg')
img2 = cv2.imread('/content/exT3_02.jpg')

# Crop the images to size 190x155
img1_cropped = img1[:155, :190]
img2_cropped = img2[:155, :190]

# Invert the pixel values using bitwise NOT
inv_img1 = cv2.bitwise_not(img1_cropped)
inv_img2 = cv2.bitwise_not(img2_cropped)

# Display the cropped and inverted images
hao.figure(figsize=(8, 4))

hao.subplot(1, 4, 1)
hao.imshow(cv2.cvtColor(img1_cropped, cv2.COLOR_BGR2RGB))
hao.title('Cropped Image 1')

hao.subplot(1, 4, 2)
hao.imshow(cv2.cvtColor(inv_img1, cv2.COLOR_BGR2RGB))
hao.title('Inverted Image 1')

hao.subplot(1, 4, 3)
hao.imshow(cv2.cvtColor(img2_cropped, cv2.COLOR_BGR2RGB))
hao.title('Cropped Image 2')

hao.subplot(1, 4, 4)
hao.imshow(cv2.cvtColor(inv_img2, cv2.COLOR_BGR2RGB))
hao.title('Inverted Image 2')

hao.tight_layout()
hao.show()
```

```python
# lab 2 cu 7 c

# Nu vn cn xut hin cc vch trng khi p dng mt n, c th c mt s nguyn nhn v cch khc phc nh sau:
#     iu chnh ngng mu: Tng hoc gim gi tr ngng mu trong khong mu bn  chn c th gip loi b cc vch trng khng mong mun. Thay i gi tr lower_red v upper_red  xem c s ci thin khng.

#     python
# lower_red = (0, 100, 100)
# upper_red = (20, 255, 255)
# Lm m nh trc khi x l: p dng mt bc lm m trc khi x l c th gip loi b nhiu v cc vch trng. Bn c th s dng b lc nh Gaussian Blur trc khi p dng mt n.
# python

#     blurred_img = cv2.GaussianBlur(img, (5, 5), 0)
#     hsv_blurred_img = cv2.cvtColor(blurred_img, cv2.COLOR_BGR2HSV)

#     Kim tra cht lng nh: m bo rng nh u vo c cht lng  tt v khng cha nhiu nhiu.

#     Kim tra mi trng nh sng: Nu nh c chp di nh sng yu hoc c nhiu nh sng, c th gy nh hng n qu trnh x l mu.

# Hy th nhng iu trn v xem liu c s ci thin khng. Nu vn gp kh khn, bn c th th nghim nhiu k thut x l nh khc nhau hoc s dng m hnh hc my  phn loi "ball".
import cv2
import matplotlib.pyplot as hao

# c nh
img1O = cv2.imread('/content/exT3_01.jpg')
img2O = cv2.imread('/content/exT3_02.jpg')

# # Ct nh
img1 = img1O[:155, 1:191]
img2 = img2O[:155, :190]

# Chuyn i sang khng gian mu HSV
hsv_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
hsv_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)

# Xc nh khong mu cho "ball" (v d: mu )
lower_red = (0, 100, 100)
upper_red = (20, 255, 255)

# To mt n (mask) t ngng mu
mask1 = cv2.inRange(hsv_img1, lower_red, upper_red)
mask2 = cv2.inRange(hsv_img2, lower_red, upper_red)

# p dng mt n ln nh  ch gi li phn ca nh c "ball"
result_img1 = cv2.bitwise_and(img1, img1, mask=mask1)
result_img2 = cv2.bitwise_and(img2, img2, mask=mask2)

# Hin th nh gc v nh ch cha "ball"
hao.figure(figsize=(10, 5))

hao.subplot(1, 2, 1)
hao.imshow(cv2.cvtColor(img1O, cv2.COLOR_BGR2RGB))
hao.title('Original Image 1')

hao.subplot(1, 2, 2)
hao.imshow(cv2.cvtColor(result_img1, cv2.COLOR_BGR2RGB))
hao.title('Image with Only "Ball" 1')

hao.tight_layout()
hao.show()

hao.figure(figsize=(10, 5))

hao.subplot(1, 2, 1)
hao.imshow(cv2.cvtColor(img2O, cv2.COLOR_BGR2RGB))
hao.title('Original Image 2')

hao.subplot(1, 2, 2)
hao.imshow(cv2.cvtColor(result_img2, cv2.COLOR_BGR2RGB))
hao.title('Image with Only "Ball" 2')

hao.tight_layout()
hao.show()

```
