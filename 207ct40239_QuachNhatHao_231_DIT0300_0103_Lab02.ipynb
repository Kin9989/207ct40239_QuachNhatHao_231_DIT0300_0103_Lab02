```python
# câu 4 lab 2
import cv2
import matplotlib.pyplot as hao

# Đọc hai ảnh
image1_path = "/content/Screenshot 2023-11-05 194917.png"
image2_path = "/content/Screenshot 2023-11-05 194922.png"

img1 = cv2.imread(image1_path)
img2 = cv2.imread(image2_path)

# Áp dụng các bộ lọc
kernel_size = 5

# Box Filter
box_filtered_img1 = cv2.boxFilter(img1, -1, (kernel_size, kernel_size))
box_filtered_img2 = cv2.boxFilter(img2, -1, (kernel_size, kernel_size))

# Gaussian Filter
gaussian_blur_img1 = cv2.GaussianBlur(img1, (kernel_size, kernel_size), 0)
gaussian_blur_img2 = cv2.GaussianBlur(img2, (kernel_size, kernel_size), 0)

# Median Filter
median_blur_img1 = cv2.medianBlur(img1, kernel_size)
median_blur_img2 = cv2.medianBlur(img2, kernel_size)

# Biểu diễn ảnh gốc và các ảnh làm mờ
hao.figure(figsize=(12, 8))

hao.subplot(2, 3, 1)
hao.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
hao.title('Ảnh Gốc 1')

hao.subplot(2, 3, 2)
hao.imshow(cv2.cvtColor(box_filtered_img1, cv2.COLOR_BGR2RGB))
hao.title('Box Filter 1')

hao.subplot(2, 3, 3)
hao.imshow(cv2.cvtColor(gaussian_blur_img1, cv2.COLOR_BGR2RGB))
hao.title('Gaussian Blur 1')

hao.subplot(2, 3, 4)
hao.imshow(cv2.cvtColor(median_blur_img1, cv2.COLOR_BGR2RGB))
hao.title('Median Blur 1')

hao.subplot(2, 3, 5)
hao.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
hao.title('Ảnh Gốc 2')

hao.subplot(2, 3, 2)
hao.imshow(cv2.cvtColor(box_filtered_img2, cv2.COLOR_BGR2RGB))
hao.title('Box Filter 2')

hao.subplot(2, 3, 3)
hao.imshow(cv2.cvtColor(gaussian_blur_img2, cv2.COLOR_BGR2RGB))
hao.title('Gaussian Blur 2')

hao.subplot(2, 3, 6)
hao.imshow(cv2.cvtColor(median_blur_img2, cv2.COLOR_BGR2RGB))
hao.title('Median Blur 2')

hao.show()
```

```python
# câu 5 lab 2
# Các kích thước bộ lọc (kernel size) đóng vai trò quan trọng trong việc xử lý ảnh và làm mờ hình ảnh. Dưới đây là một số nhận xét về các kích thước filter:

#     1 Kích Thước Bộ Lọc Lớn:
#         Khi sử dụng bộ lọc lớn (ví dụ: kernel size lớn), hiệu ứng làm mờ sẽ mạnh mẽ hơn và làm giảm chi tiết trong hình ảnh.
#         Tuy nhiên, điều này có thể làm mất thông tin quan trọng nếu kích thước quá lớn.

#     2 Kích Thước Bộ Lọc Nhỏ:
#         Bộ lọc nhỏ (ví dụ: kernel size nhỏ) sẽ giữ lại nhiều chi tiết hơn trong hình ảnh, nhưng cũng làm giảm độ mờ.
#         Có thể tạo ra hiệu ứng làm mờ nhẹ và giữ lại nhiều chi tiết cần thiết.

#     3 Kernel Size Là Số Lẻ:
#         Trong một số trường hợp, đặc biệt là khi sử dụng median filter, việc chọn kernel size là số lẻ thường được khuyến khích. Điều này giúp giữ lại giá trị trung bình của pixel một cách hiệu quả hơn.

#     4 Lựa Chọn Kernel Size Phù Hợp:
#         Lựa chọn kích thước bộ lọc phụ thuộc vào đặc tính của hình ảnh và mục tiêu của quá trình xử lý.
#         Cần thử nghiệm và điều chỉnh kích thước bộ lọc để đạt được hiệu quả tốt nhất cho từng tình huống cụ thể.

#     5 Tùy Chọn Tối Ưu:
#         Trong một số trường hợp, sự kết hợp của các bộ lọc khác nhau hoặc sự thay đổi kích thước bộ lọc theo từng giai đoạn có thể dẫn đến kết quả tốt hơn.
```

```python
# câu 6 lab 2
import cv2
import matplotlib.pyplot as hao

# Đọc ảnh và làm mờ
img = cv2.imread('/content/pepper_noise.jpg')
blur = cv2.medianBlur(img, 5)

# Chuyển ảnh trắng đen sang ảnh màu
color_img = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)
# Chuyển ảnh trắng đen sang ảnh màu xanh
enhanced_green_img = color_img.copy()
enhanced_green_img[:, :, 1] = cv2.equalizeHist(color_img[:, :, 1])

# Chuyển ảnh màu sang ảnh đen trắng
gray_img = cv2.cvtColor(color_img, cv2.COLOR_RGB2GRAY)

# Cân bằng sáng với equalizeHist
equalized_img = cv2.equalizeHist(gray_img)

# Biểu diễn các ảnh
hao.figure(figsize=(10, 8))

hao.subplot(2, 2, 1)
hao.imshow(enhanced_green_img)
hao.title('Ảnh Màu Xanh')

hao.subplot(2, 2, 2)
hao.imshow(gray_img, cmap='gray')
hao.title('Ảnh Đen Trắng')

hao.subplot(2, 2, 3)
hao.imshow(equalized_img, cmap='gray')
hao.title('Ảnh Cân Bằng Sáng')

hao.show()
```

```python
import cv2
import matplotlib.pyplot as hao

# Read the two images
img1 = cv2.imread('/content/exT3_01.jpg')
img2 = cv2.imread('/content/exT3_02.jpg')

# Crop the images to size 190x155
img1_cropped = img1[:155, :190]
img2_cropped = img2[:155, :190]

# Invert the pixel values using bitwise NOT
inv_img1 = cv2.bitwise_not(img1_cropped)
inv_img2 = cv2.bitwise_not(img2_cropped)

# Display the cropped and inverted images
hao.figure(figsize=(8, 4))

hao.subplot(1, 4, 1)
hao.imshow(cv2.cvtColor(img1_cropped, cv2.COLOR_BGR2RGB))
hao.title('Cropped Image 1')

hao.subplot(1, 4, 2)
hao.imshow(cv2.cvtColor(inv_img1, cv2.COLOR_BGR2RGB))
hao.title('Inverted Image 1')

hao.subplot(1, 4, 3)
hao.imshow(cv2.cvtColor(img2_cropped, cv2.COLOR_BGR2RGB))
hao.title('Cropped Image 2')

hao.subplot(1, 4, 4)
hao.imshow(cv2.cvtColor(inv_img2, cv2.COLOR_BGR2RGB))
hao.title('Inverted Image 2')

hao.tight_layout()
hao.show()
```

```python
# lab 2 câu 7 c

# Nếu vẫn còn xuất hiện các vạch trắng khi áp dụng mặt nạ, có thể có một số nguyên nhân và cách khắc phục như sau:
#     Điều chỉnh ngưỡng màu: Tăng hoặc giảm giá trị ngưỡng màu trong khoảng màu bạn đã chọn có thể giúp loại bỏ các vạch trắng không mong muốn. Thay đổi giá trị lower_red và upper_red để xem có sự cải thiện không.

#     python
# lower_red = (0, 100, 100)
# upper_red = (20, 255, 255)
# Làm mờ ảnh trước khi xử lý: Áp dụng một bước làm mờ trước khi xử lý có thể giúp loại bỏ nhiễu và các vạch trắng. Bạn có thể sử dụng bộ lọc như Gaussian Blur trước khi áp dụng mặt nạ.
# python

#     blurred_img = cv2.GaussianBlur(img, (5, 5), 0)
#     hsv_blurred_img = cv2.cvtColor(blurred_img, cv2.COLOR_BGR2HSV)

#     Kiểm tra chất lượng ảnh: Đảm bảo rằng ảnh đầu vào có chất lượng đủ tốt và không chứa nhiễu nhiều.

#     Kiểm tra môi trường ánh sáng: Nếu ảnh được chụp dưới ánh sáng yếu hoặc có nhiễu ánh sáng, có thể gây ảnh hưởng đến quá trình xử lý màu.

# Hãy thử những điều trên và xem liệu có sự cải thiện không. Nếu vẫn gặp khó khăn, bạn có thể thử nghiệm nhiều kỹ thuật xử lý ảnh khác nhau hoặc sử dụng mô hình học máy để phân loại "ball".
import cv2
import matplotlib.pyplot as hao

# Đọc ảnh
img1O = cv2.imread('/content/exT3_01.jpg')
img2O = cv2.imread('/content/exT3_02.jpg')

# # Cắt ảnh
img1 = img1O[:155, 1:191]
img2 = img2O[:155, :190]

# Chuyển đổi sang không gian màu HSV
hsv_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
hsv_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)

# Xác định khoảng màu cho "ball" (ví dụ: màu đỏ)
lower_red = (0, 100, 100)
upper_red = (20, 255, 255)

# Tạo mặt nạ (mask) từ ngưỡng màu
mask1 = cv2.inRange(hsv_img1, lower_red, upper_red)
mask2 = cv2.inRange(hsv_img2, lower_red, upper_red)

# Áp dụng mặt nạ lên ảnh để chỉ giữ lại phần của ảnh có "ball"
result_img1 = cv2.bitwise_and(img1, img1, mask=mask1)
result_img2 = cv2.bitwise_and(img2, img2, mask=mask2)

# Hiển thị ảnh gốc và ảnh chỉ chứa "ball"
hao.figure(figsize=(10, 5))

hao.subplot(1, 2, 1)
hao.imshow(cv2.cvtColor(img1O, cv2.COLOR_BGR2RGB))
hao.title('Original Image 1')

hao.subplot(1, 2, 2)
hao.imshow(cv2.cvtColor(result_img1, cv2.COLOR_BGR2RGB))
hao.title('Image with Only "Ball" 1')

hao.tight_layout()
hao.show()

hao.figure(figsize=(10, 5))

hao.subplot(1, 2, 1)
hao.imshow(cv2.cvtColor(img2O, cv2.COLOR_BGR2RGB))
hao.title('Original Image 2')

hao.subplot(1, 2, 2)
hao.imshow(cv2.cvtColor(result_img2, cv2.COLOR_BGR2RGB))
hao.title('Image with Only "Ball" 2')

hao.tight_layout()
hao.show()

```
